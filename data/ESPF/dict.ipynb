```python
import pandas as pd
import numpy as np
from subword_nmt.apply_bpe import BPE
import codecs

# Data folder and file paths
dataFolder = 'info'
smiles_file = '../unique_drugs_with_smiles.csv'  # Assuming the filename is 'smiles.csv'

# Read the CSV file
smiles_df = pd.read_csv(smiles_file)
smiles_list = smiles_df['smiles'].values  # Extract all SMILES data

# Load BPE (Byte Pair Encoding) model
# For Drugs
vocab_path = dataFolder + '/codes_drug_chembl_1500.txt'
bpe_codes_drug = codecs.open(vocab_path)
dbpe = BPE(bpe_codes_drug, merges=-1, separator='')
sub_csv = pd.read_csv(dataFolder + '/subword_units_map_drug_chembl_1500.csv')

idx2word_d = sub_csv['index'].values
words2idx_d = dict(zip(idx2word_d, range(0, len(idx2word_d))))

# Initialize dictionary to store SMILES and their corresponding vectors
smiles_vectors = {}

# Convert each SMILES into a vector and store in the dictionary
for smile in smiles_list:
    t = dbpe.process_line(smile).split()
    i = [words2idx_d[i] for i in t]
    v = np.zeros(len(idx2word_d), )
    v[i] = 1

    # vector = np.array(v)  # Convert to numpy array
    vector = np.array(i)
    smiles_vectors[smile] = vector

# Save the dictionary as a .npy file
npy_file_path = 'ESPF_smiles_vectors.npy'
np.save(npy_file_path, smiles_vectors)

print(f"SMILES and corresponding vectors have been saved to {npy_file_path}")
```
